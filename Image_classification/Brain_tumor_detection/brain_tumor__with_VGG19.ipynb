```python
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Dense, Flatten
from keras.models import Model,load_model
from keras.applications.vgg19 import VGG19 , preprocess_input
import keras
```

```python
%matplotlib inline
```

Model building

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
baseModel=VGG19(input_shape=(224,224,3),include_top=False)
```

Preparing Data useing Data Generator

```python
def preprocessingImages_train(path):
  """
  input : path
  output: Pre processed images
  """
  image_data = ImageDataGenerator(zoom_range= 0.2, shear_range=0.2, preprocessing_function=preprocess_input, horizontal_flip=True)
  image = image_data.flow_from_directory(directory=path , target_size=(224,224), batch_size=32, class_mode='binary')
  return image
```

```python
path = "/content/drive/MyDrive/Colab_Notebooks/project1/brain_train_data/train"
train_data = preprocessingImages_train(path)
```

```python
def preprocessingImages_test(path):
  """
  input : path
  output: Pre processed images
  """
  image_data = ImageDataGenerator(preprocessing_function=preprocess_input)
  image = image_data.flow_from_directory(directory=path , target_size=(224,224), batch_size=32, class_mode='binary')
  return image
```

```python
path = "/content/drive/MyDrive/Colab_Notebooks/project1/brain_train_data/test"
test_data = preprocessingImages_test(path)
```

```python
path = "/content/drive/MyDrive/Colab_Notebooks/project1/brain_train_data/validation"
val_data = preprocessingImages_test(path)
```

```python
for layer in baseModel.layers:
  layer.trainable=False
```

```python
#baseModel.summary()
X=Flatten()(baseModel.output)
X=Dense(units=1,activation='sigmoid')(X)
model=Model(baseModel.input,X)
```

```python
#model.summary()
model.compile(optimizer='adam',loss=keras.losses.binary_crossentropy, metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
mc = ModelCheckpoint(filepath="/content/drive/MyDrive/Colab_Notebooks/project1/1.brain-tumor/brain_tumor_VGG.h6", monitor='val_accuracy',verbose=1,save_best_only=True)
es = EarlyStopping(monitor='val_accuracy',min_delta=0.01,patience=6,verbose=1)
cb=[mc,es]
```

```python
hist=model.fit_generator(train_data,steps_per_epoch=16, epochs=30,validation_data=val_data, callbacks=cb)
```

```python
model = load_model("/content/drive/MyDrive/Colab_Notebooks/project1/1.brain-tumor/brain_tumor_VGG.h6")
```

```python
acc = model.evaluate_generator(test_data)[1]
print(f"Accuracy={100*acc}%")
```

```python
h=hist.history
h.keys()
```

```python
plt.plot(h['accuracy'])
plt.plot(h['val_accuracy'] , c= 'red')

plt.title("acc vs val-acc")
```

```python
plt.plot(h['loss'])
plt.plot(h['val_loss'] , c= 'red')

plt.title("acc vs val-acc")
```
